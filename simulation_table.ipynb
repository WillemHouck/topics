```python
import pandas as pd

data = pd.read_csv("simulation_results.csv").drop("Unnamed: 0", axis=1)
data
```

Preprocessing

```python
plugin = pd.concat([data.iloc[:, :2], data.iloc[:, 6:]], axis=1)
plugin["model"] = "Plug-in"
plugin = plugin.rename(columns={"plug_in_slope": "slope", "plug_in_intercept":"intercept"})

lts = pd.concat([data.iloc[:, 2:4], data.iloc[:, 6:]], axis=1)
lts["model"] = "LTS"
lts = lts.rename(columns={"lts_slope": "slope", "lts_intercept":"intercept"})

ols = data.iloc[:, 4:]
ols["model"] = "OLS"
ols = ols.rename(columns={"ols_slope": "slope", "ols_intercept":"intercept"})

data = pd.concat([plugin, lts, ols])

data
```

```python
data["config"] = None

data.loc[(data.bl==0) & (data.vo==0), ["config"]] = "C1"
data.loc[(data.bl==0.15) & (data.vo==0), ["config"]] = "C2"
data.loc[(data.bl==0) & (data.vo==0.15), ["config"]] = "C3"
data.loc[(data.bl==0.1) & (data.vo==0.1), ["config"]] = "C4"

display(data)
```

```python
slope = pd.concat([data.iloc[:, :1], data.iloc[:, 2:]], axis=1)
slope["type"] = "Slope"
slope = slope.rename(columns={"slope": "value"})

intercept = data.iloc[:, 1:]
intercept["type"] = "Intercept"
intercept = intercept.rename(columns={"intercept": "value"})

data = pd.concat([slope,intercept])
```

```python
table = data.pivot_table(index=["n","model"], columns=["config","type"], values="value")

display(table)
print(table.to_latex())
```
